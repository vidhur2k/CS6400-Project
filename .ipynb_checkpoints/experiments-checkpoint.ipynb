# Experiments Comparing the Performance of MongoDB and PostgreSQL

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Experiment 1: Latency

### 1.1 Read Latency Over the Number of Rows

```python
# Query: Select * from teams LIMIT X
execution_times = np.array([0.068, 0.224, 0.339, 0.585, 0.609])
m_execution_times = np.array([0.006, 0.048, 0.063, 0.071, 0.094])
rows = np.array([100, 500, 1000, 1500, 2280])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.plot(rows, execution_times, label='PostgreSQL')
plt.plot(rows, m_execution_times, label='MongoDB', c='green')
plt.xlabel('Number of Rows')
plt.ylabel('Latency (ms)')
plt.legend()
plt.title('Query Execution Time for selecting all EPL games')
plt.show()
```
